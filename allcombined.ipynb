```python
import xarray as xr
import numpy as np
import glob
from netCDF4 import Dataset
import os
```

```python
file_path='/users/karolina/Desktop/AREX_netCDF_1996-2023/*/netcdf/*.nc' 
file_list = glob.glob(file_path)
```

```python
datasets=[xr.open_dataset(fp) for fp in file_list]
print(len(datasets))
```

```python
output_path = '/users/karolina/Desktop/AREX_netCDF_1996-2023/combined_all.nc'

# Empty netcdf file:
rootgrp = Dataset(output_path, 'w')

for fp in file_list:
    # name of the group = name of the file
    group_name = os.path.splitext(os.path.basename(fp))[0]
    ds = xr.open_dataset(fp)
    # Saving dataset to group
    ds.to_netcdf(output_path, mode='a', group=group_name)

rootgrp.close()
```

```python
root = Dataset(output_path, 'r')
print("Grupy w pliku:", list(root.groups.keys()))
```

```python
aligned_datasets=[
    ds.pad(n_levels=(0, max_levels-ds.sizes["n_levels"]), constant_values=np.nan)
    if "n_levels" in ds.sizes else ds
    for ds in datasets
]
print(aligned_datasets[:5])
```
